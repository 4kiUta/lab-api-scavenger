# Notes on best practice on how to compare dates in python 


Python compares the strings **lexicographicaly** and in this case it works, but its not the right way to go, because it can also accepts the wrong dates!
Here is a Counterexample:

>>> a ='2009-33-10'

>>> b ='2009-11-1'

>>> a>b

>>>True

As a proper way for dealing with dates you should use datetimemodule which has a lot of tools for working with date objects.
You can convert your strings to date object with datetime.datetime.strptime and then you can use basic arithmetic operation to compare your date objects, as theyve been supported already by this module.
